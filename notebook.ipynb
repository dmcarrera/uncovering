What characteristics enable a business to stand the test of time?

To help answer this question, BusinessFinancing.co.uk researched the oldest company still in business in **almost** every country and compiled the results into several CSV files. This dataset has been cleaned.

Tener información útil en diferentes archivos es un problema común. Aunque es mejor mantener diferentes tipos de datos separados para el almacenamiento de datos, querrás tener todos los datos en un solo lugar para el análisis. Utilizarás la unión y la manipulación de datos para trabajar con esta información y comprender mejor las empresas más antiguas del mundo.

```python
# Import necessary libraries
import pandas as pd

# Load the data
businesses = pd.read_csv("data/businesses.csv")
new_businesses = pd.read_csv("data/new_businesses.csv")
countries = pd.read_csv("data/countries.csv")
categories = pd.read_csv("data/categories.csv")
```

```python
# Start coding here!
# Use as many cells as you like

print(businesses.info())
print('---')
print(new_businesses.info())
print('---')
print(countries.info())
print('---')
print(categories.info())

```

```python
#oldest_business_continent

#combinar negocios y paises

business_continent = businesses.merge(countries, on='country_code')

continent = business_continent.groupby('continent').agg({"year_founded":"min"})
continent.head()
```

```python
#con cuatro columnas: continente, país, negocio y año_fundado
merge_continent = business_continent.merge(continent, on=['continent', 'year_founded'] )
#cambia el orden de las columnas
oldest_business_continent = merge_continent[['continent', 'country', 'business',  'year_founded']]
oldest_business_continent
```

¿Cuántos países por continente carecen de datos sobre los negocios más antiguos? ¿Cambiaría esto al incluir nuevos negocios? Cuenta el número de países por continente que faltan datos de negocios, incluyendo los nuevos negocios, y guarda los resultados en un DataFrame llamado `count_missing` con columnas para el continente y el conteo.

```python
#unir nuevos y viejos

all_bussines = pd.concat([businesses, new_businesses])


# Realice una nueva fusión entre los datos de empresas y países. Utilice parámetros 
# adicionales esta vez para realizar una fusión externa y crear una columna de indicador 
# para ver mejor los valores faltantes. Una fusión externa combina dos DataFrames en función de una columna
#  clave e incluye todas las filas de ambos DataFrames.
new_all_countries = all_bussines.merge(countries, on='country_code', how='outer', indicator=True)

#count_missing = continent_completed.groupby('continent')['business']
new_missing_countries = new_all_countries[new_all_countries['_merge']!= 'both']

count_missing = new_missing_countries.groupby('continent').agg({'country':'count'})

count_missing.columns = ['count_missing'] 
```

```python
b_categories = businesses.merge(categories, on='category_code')

b_categories_country = b_categories.merge(countries, on='country_code')
```
